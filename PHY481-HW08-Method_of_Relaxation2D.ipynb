```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

In this homework assignment I had to implement the method of relaxation, a method which is used to calculate the electric potential (in volts V) in a specific area of space, in two dimensions. This code simply takes in the initial conditions on each boundary of this square and averages out the values across the entire array.

```python
# Number of iterations
maxIter = 500
# Set Dimension and delta
lenX = lenY = 30
delta = 1
# Boundary conditions
Vtop = 0
Vbottom = 0
Vleft = 10
Vright = -10
# Initial grid value
Vinit = 0
# Setting hte color to make it look nice
colorinterpolation = 50
colorMap = plt.cm.jet
# Set meshgrid
X, Y = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY))
# Setting array size and default value for the grid
V = np.empty((lenX, lenY))
V.fill(Vinit)
# Boundary conditions
V[(lenY-1):, :] = Vtop
V[:1, :] = Vbottom
V[:, (lenX-1):] = Vright
V[:, :1] = Vleft
# Loop that will calculate the average of all surrounding points
for iteration in range(0, maxIter):
    for i in range(1, lenX-1):
        for j in range(1, lenY-1):
            V[i, j] = (V[i+1][j] + V[i-1][j] + V[i][j+1] + V[i][j-1])/4
# Creating the contour map
plt.title("Contour of Potential")
plt.contourf(X, Y, V, colorinterpolation, cmap=colorMap)
plt.colorbar()
```

```python
# Simpler Contour Plot
CS = plt.contour(X,Y,V,50,cmap=colorMap)
plt.colorbar()
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Contour Map of Potential')
```

```python
# 3D Plot of the Potential
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, V, cmap=colorMap)
plt.title('Electric Potential Across the Square')
```
