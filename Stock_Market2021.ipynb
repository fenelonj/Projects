```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
import datetime
import math
import time
from IPython.display import clear_output
```

```python
NDAQ = yf.download(tickers='NDAQ', period='1d', interval='5m')
NDAQ = NDAQ.replace(np.nan, 0)
NDAQ = NDAQ.reset_index()
for i in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
    NDAQ[i] = NDAQ[i].astype('float64')
if NDAQ['Open'].iloc[-1] > NDAQ['Open'].iloc[0]:
    color = 'limegreen'
if NDAQ['Open'].iloc[-1] < NDAQ['Open'].iloc[0]:
    color = 'red'
```

```python
NDAQ.plot(x='Datetime', y='Open', figsize=(7,3), grid=True, title='Price of NDAQ', c=color)
```

```python
def getBTC():
    i = yf.download(tickers='BTC-USD', period='1d', interval='1m')
    return i
fig = plt.figure(figsize=(7,4))
price = []
datetime = []
while True:
    BTC = getBTC()
    BTC = BTC.reset_index()
    for i in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
        BTC[i] = BTC[i].astype('float64')
    print('Current Price of BTC is:', BTC['Open'].iloc[-1].round(2))
    growth = ((BTC['Open'].iloc[-1]-BTC['Open'].iloc[0])*100/BTC['Open'].iloc[0]).round(3)
    if growth > 0:
        print('BTC is up:', growth, '%')
    if growth < 0:
        print('BTC is down:', growth, '%')
    price.append(BTC['Open'].iloc[-1])
    datetime.append(BTC['Datetime'].iloc[-1])
    plt.plot(datetime, price)
    #BTC.plot(x='Datetime', y='Open', title='Price of Bitcoin',figsize=(10,4))
    plt.show()
    time.sleep(60)
    clear_output(wait=True)
    display(fig)
    fig.clear()
```

```python
SOXL = yf.download(tickers='SOXL', period='1d', interval='1m')
SOXL = SOXL.reset_index()
for i in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
    SOXL[i] = SOXL[i].astype('float64')

if SOXL['Open'].iloc[-1] < SOXL['Open'].iloc[0]:
    color = 'red'
if SOXL['Open'].iloc[-1] > SOXL['Open'].iloc[0]:
    color = 'limegreen'
SOXL.plot(x='Datetime', y='Open', figsize=(7,3), c=color, title = 'Price of SOXL ETF', grid=True)
plt.style.use('default')
growth = (((SOXL['Open'].iloc[-1]-SOXL['Open'].iloc[0])/SOXL['Open'].iloc[0])*100).round(3)
print('Todays growth of SOXL:', growth, '%')
```
